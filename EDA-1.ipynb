```python
# Checking missing values  
file.isnull().sum()
```

```python
# dropping the missing values 
file.dropna(inplace=True)
file
```

```python
#Checking is there any missing values are left 
file.isnull().sum()
```

```python
# Duplicated values 
file.duplicated().sum()
```

```python
file=file.drop_duplicates(ignore_index=True)
file
```

```python
# checking is there any Duplicated values 
file.duplicated().sum()
```

```python
# using Histogram for data visualize
import matplotlib.pyplot as plt
import seaborn as sns 
plt.figure(figsize=(10,8))
sns.histplot(data=file,bins=30)
plt.title("Data distributions of Cardiographic")
```

```python
# find the outliers
Q1=file.quantile(0.25)
Q1
```

```python
Q3=file.quantile(0.75)
Q3
```

```python
IQR=Q3-Q1
IQR
```

```python
U_E=Q3+1.5*IQR
U_E
```

```python
L_E=Q1-1.5*IQR
L_E
```

```python
Outliers=(file<L_E) | (file>U_E)
print(Outliers.any(axis=1))
```

```python
# remove Outliers
Cleaned_file=file[~Outliers]
Cleaned_file.dropna(inplace=True)
```

```python
No_Outliers=(file>L_E) | (file<U_E)
print(No_Outliers.any(axis=1))
```

```python
# Plot the boxplot before removing outliers
plt.figure(figsize=(12,8))
sns.boxplot(data=Outliers)
plt.show()
```

```python
# Plot the boxplot after removing outliers
plt.figure(figsize=(12,8))
sns.boxplot(data=No_Outliers)
plt.show()
```
