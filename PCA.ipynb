# Dimensionality Reduction with PCA

```python
# Standardize the features 
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaled_data=scaler.fit_transform(wine_data)
```

```python
# Implement PCA
from sklearn.decomposition import PCA

# Initialize the PCA model 
pca_model=PCA(n_components=4)
pca_data=pca_model.fit_transform(scaled_data)

# Convert this pca data into dataframe 
PCA_2D=pd.DataFrame(pca_data)

# Display the PCA_explained variance ratio
print(pca_model.explained_variance_ratio_)
```

```python
# Use a scree plot and cumulative explained variance to determine the optimal number of principal component
plt.figure(figsize=(6,6))
plt.plot(range(1,len(pca_model.explained_variance_ratio_)+1),pca_model.explained_variance_ratio_,marker='o',linestyle="--")
plt.title("ScreePlot")
plt.xlabel("Principal Component")
plt.ylabel("Explained Variance Ratio")
plt.show()
```

```python
# Cumulative explained variance
cumulative_explained_variance = pca_model.explained_variance_ratio_.cumsum()
plt.figure(figsize=(6, 6))
plt.plot(range(1,len(cumulative_explained_variance)+1),cumulative_explained_variance, marker='o', linestyle='--')
plt.title('Explained ratio by components')
plt.xlabel('Number of components')
plt.ylabel('Cumulative Explained Variance')
plt.show()
```
